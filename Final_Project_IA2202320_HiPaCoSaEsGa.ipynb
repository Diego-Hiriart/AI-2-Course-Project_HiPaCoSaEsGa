## Ingeniería de características
- Existen dos objetivos principales de realizar este proceso:
    - Obtener datos de las muestras que sean representativos de las clases del dataset, es decir, valores comunes a todos los miembros de un dígito que permiten diferenciarlos.
    - Reducir el tamaño de cada muestra usada para entrenamiento. Al usar solo características representativas, se reduce el tamaño de los inputs al modelo de inteligencia artificial, mejorando el rendimiento al entrenarlos y al usarlos para inferencia/predicción.

### 1. Características basada en estadísticas y matemática
- Las primeras estadísticas que se extraen consisten en obtener datos estadísticos en base a los pixeles de la imagen. Se utiliza el dataset de muestras binarizadas para este propósito.

```python
def skeletonize_digit(digit: np.ndarray) -> np.ndarray:
    """
    Skeletonize a digit.
    Returns the skeletonized digit.
    """
    return skeletonize(digit).reshape(-1)


def extract_stats_features(dataset: pd.DataFrame, skeletonized: bool = False) -> pd.DataFrame:
    """
    Extract statistical features from dataset.
    Extracted features are: std and mean.
    Returns a dataframe with the extracted features.
    """
    features = pd.DataFrame()
    # Iterate over each row
    for index, row in dataset.iterrows():
        X = row.drop('label')
        reshaped_X = X.values.reshape(28,28)
        y = row['label']
        
        if skeletonized:
            X = skeletonize_digit(reshaped_X)

        # Get features
        features.loc[index, 'label'] = y
        features.loc[index, 'std'] = X.std()
        features.loc[index, 'mean'] = X.mean()
        width_ds = np.where(reshaped_X.sum(axis=0))[0] # Get indexes of columns with at least one pixel
        features.loc[index, 'width'] = width_ds[-1] - width_ds[0]
        height_ds = np.where(reshaped_X.sum(axis=1))[0] # Get indexes of rows with at least one pixel
        features.loc[index, 'height'] = height_ds[-1] - height_ds[0]
    features['label'] = features['label'].astype(int)
    return features
```

    - Ancho: el ancho del número contado en pixeles, es decir, desde el primer hasta el último pixel lleno de forma vertical.
    - Alto: el alto del número contado en pixeles, es decir, desde el primer hasta el último pixel lleno de forma horizontal.

```python
def extract_pixel_count(dataset: pd.DataFrame) -> pd.DataFrame:
    """
    Count the number of pixels per row and column and concatenate them.
    Extracted features are the sum of pixels per row and column.
    Returns a dataframe with the extracted features.
    """
    features = pd.DataFrame()
    # Iterate over each row
    for index, row in dataset.iterrows():
        # Process row
        X = row.drop('label') # Drop label
        X[X > 0] = 1 # Binarize
        X = X.values.reshape(28,28) # Reshape
        y = row['label']
        # Get features
        sum_row = X.sum(axis=0)
        sum_col = X.sum(axis=1)
        concatenated_sum = np.concatenate((sum_row, sum_col))
        features.loc[index, 'label'] = y
        # Concatenate all features
        feature_labels = [f'sum{i}' for i in range(len(concatenated_sum))]
        features.loc[index, feature_labels] = concatenated_sum
    features = features.astype(int)
    return features
```

## 4. Uso de pipelines para exploración de modelos de clasificación y regresión
- Para explorar el desempeño de los modelos de clasificación y regresión con los distintos datasets generados, se debe realizar una exploración de varias configuraciones de entrenamiento de los modelos de inteligencia artificial. 
- Entrenar modelos puede tomar largo tiempo, más aun cuando se trata de una exploración para encontrar las mejores combinaciones de datos e hiperparámetros de entrenamiento. Cuando se habla de hiperparámetros de entrenamiento, se hace referencia a configuraciones que pueden cambiarse de los modelos para afectar la manera en al que aprender y obtener mejores resultados. 
- Considerando esto, se utiliza la herramienta de pipelines de la librería scikit-learn, que permite configurar el entrenamiento de múltiples modelos para ejecutarlos en grupo y obtener los resultados de desempeño de todos. Estos pipelines permiten definir pasos que deben seguirse, modelos que se deben entrenar, combinaciones de hiperparámetros que probar, y los datos que se deben utilizar para su ejecución.
- A este pipeline se le indica que debe realizar un "grid search", que es la exploración de hiperparámetros para varios modelos, para cada dataset.

**Selección de parámetros a explorar**
- El ajuste de hiperparámetros que se explorará para MLP se basa en el siguiente razonamiento (Ansari, 2022):
    - Learning Rate (velocidad de aprendizaje): establece que tanto se deben cambiar los parámetros, se deben explorar valores de learning rate desde pequeños a grandes. Esto permite encontrar el rango de LRs que no son demasiado grandes como para no llegar al loss mínimo y no muy pequeños como para necesitar demasiadas épocas de entrenamiento.
    - Batch size (tamaño de muestra): indica después de cuantos samples se debe recalcular el loss y ajustar pesos, si es muy pequeño los ajustes de dan más seguido y mejora el accuracy pero se requiere mucho más poder computacional. Se debe encontrar el valor que permite un buen accuracy pero no demande demasiados recursos y sea ineficiente.
    - Hidden neurons (neuronas escondidas): la cantidad de neuronas afecta que tan complejo es un modelo, si un modelo es demasiado simple no es capaz de aprender patrones para funcionar correctamente. Pero se debe cuidar que la complejidad no sea tan alta que consuma mucho tiempo entrenar o exista overfitting, esto último es cuando el modelo aprende mucho de los datos de entrenamiento y se vuelve incapaz de generalizar para funcionar adecuadamente con datos reales o de prueba.
    - Max iter (pasos de entrenamiento): indica cuantas veces el modelo debe entrenar con los datos, es decir, la cantidad de ajustes. Si se hacen demasiados se puede gastar mucho tiempo o hasta tener overfitting, si se hacen muy pocos el modelo no aprende lo suficiente para tener buen desempeño.


- Respecto a la exploración de hiperparámetros para Random Forest, se tiene consideración los siguientes efectos de los hiperparámetros en dicho tipo de modelo (Arya, 2022):
    - Number of estimators (número de estimadores): la cantidad de árboles en el Random Forest, con más árboles existe más precisión, pero una cantidad demasiado grande es ineficiente para tiempo de entrenamiento.
    - Max depth (máxima profundidad): se refiere a que tantas divisiones (decisiones) puede tener cada árbol en el modelo, si es muy bajo no hay suficiente aprendizaje, llamado underfitting (Ram, 2020).
    - Min samples split (mínima cantidad de muestras para división): especifica cuantas muestras debe tener un nodo del árbol para dividirse y tomar una decisión, se busca una cantidad de muestras que le permita al árbol lograr suficientes divisiones para evitar underfitting.
    
- Para ambas exploraciones, no existe un criterio específico que permita saber qué valores usar, se deben explorar valores que se sepa permitan mejor rendimiento; por ejemplo, más árboles en Random Forest, o más iteraciones en MLP.
- Se genera un conjunto de los 5 datasets a utilizar, el original en escala de grises, el binarizado, las características estadísticas/matemáticas, conteo de pixeles, y combinación de estadísticas/matemáticas con pixeles.
- La métrica que usará para evaluar los modelos, al ser este un problema de clasificación, es _accuracy_ o exactitud/precisión. Esta métrica indica cuantas muestras de clasificaron correctamente (Mishra, 2018).
    - Para los modelos de clasificación, solo se calcula la cantidad de clasificaciones correctas.
    - En los modelos de regresión, se toma el escalar resultante de la regresión y se lo redondea al entero más cercano. Este entero indica a que dígito pertenece la muestra. En caso de resultados menores a cero o mayores a nueve, se los redondea a 0 y 9 respectivamente.

```python
# Define the parameter grid for hyperparameter tuning
param_grid = [
    {
        'estimator': [RandomForestClassifier()],
        'estimator__n_estimators': [100, 200],
        'estimator__max_depth': [5, 10, 15],
        'estimator__min_samples_split': [2],
    },
    {
        'estimator': [RandomForestRegressor()],
         'estimator__n_estimators': [100, 200],
        'estimator__max_depth': [5, 10, 15],
        'estimator__min_samples_split': [2],
    },
    {
        'estimator': [MLPClassifier()],
        'estimator__learning_rate_init': [0.001, 0.01],
        'estimator__batch_size': [1024],
        'estimator__max_iter': [400, 800],
        'estimator__hidden_layer_sizes': [(100,), (50, 50)],
    },
    {
        'estimator': [MLPRegressor()],
        'estimator__learning_rate_init': [0.001, 0.01],
        'estimator__batch_size': [1024],
        'estimator__max_iter': [400, 800],
        'estimator__hidden_layer_sizes': [(100,), (50, 50)],
    }
]

def round_pred(y_pred):
    # Round the predictions to the nearest integer
    # If the predictions are outside the range [0, 9], clip them
    # This will not have any effect on classification
    convert_to_nearest = y_pred.clip(0, 9).round().astype(int)
    return convert_to_nearest

# Define the scoring metrics for classification and regression
scoring = {
    'accuracy': make_scorer(lambda y_true, y_pred: accuracy_score(y_true, round_pred(y_pred))),
}

# Define the input data types
input_types = {
    'grayscale': (
        mnist.drop(columns=['label']),
        mnist['label'],
    ),
    'binary': (
        mnist.drop(columns=['label']).applymap(lambda x: 1 if x > 0 else 0),
        mnist['label'],
    ),
    'stats': (
        skeletonized_stats_features.drop(columns=['label']),
        skeletonized_stats_features['label'],
    ),
    'pixel_count': (
        pixel_count_features.drop(columns=['label']),
        pixel_count_features['label'],
    ),
    'joined_features': (
        joined_features.drop(columns=['label']),
        joined_features['label'],
    ),
}
```

#### Entrenamiento de modelos para exploración
- Se entrenan los modelos con los datasets e hiperparámetros escogidos para determinar las mejores combinaciones de estos.
- Para realizar la validación del desempeño de los modelos, no se pueden utilizar las mismas muestras de entrenamiento, pues los modelos están aprendiendo con esto y las conoces cada vez mejor. Por lo tanto, se utiliza _k-fold Cross Validation_, esta técnica consiste en dividir el dataset en _k_ partes iguales, y usar una de estas para validar y las demás para entrenar; se puede realizar el proceso _k_ veces de modo que se use cada división una vez para validar (Brownlee, 2018).
- Se decide utilizar un valor de $k=4$ para el Cross Validation dado que permite 4 iteraciones por cada modelo a explorar (que no son demasiadas) pero mantiene un buen porcentaje de datos de entrenamiento, 75%, o 31500 muestras.
- Para agilizar los 560 entrenamiento que deben realizarse (28 combinaciones de modelos, multiplicado por 4 iteraciones de _k-fold Cross Validation_ dado que $k=4$, multiplicado por los 5 datasers existentes) se utiliza procesamiento paralelo con núcleos físicos de CPU, scikit learn no tiene soporte para GPU. Adicionalmente, se corrieron las exploraciones con los datasets en escala de grises y binarizado en un computadora y los demás en otra, para disminuir el tiempo de exploración; se guardaron los resultados en archivos con la librería de luego fueron unidos en un solo diccionario de resultados.

```python
grid_search_results = {}

for input_type, data in input_types.items():
    print(f'Input type: {input_type}')
    X, y = data

    # Define the pipeline
    pipeline = Pipeline([
        ('estimator', RandomForestClassifier())  # Estimator
    ])

    # Perform grid search with cross-validation
    grid_search = GridSearchCV(
        pipeline,
        param_grid,
        scoring=scoring,
        cv=4,
        refit=False,
        verbose=10,
        n_jobs=5, # Parallelize the grid search using many CPU cores
        return_train_score=True
    )

    grid_search.fit(X, y)

    # Save the results
    grid_search_results[input_type] = grid_search
```

# Resultados

- El siguiente código, permite preparar los datos para visualización de todos los resultados.
- Se utilizan los datos generados por la exploración y se añade a que dataset y tipo de modelo (clasificación o regresión) pertenece ese entrenamiento en específico.

```python
estimator_types = {
    'classification': ['RandomForestClassifier', 'MLPClassifier'],
    'regression': ['RandomForestRegressor', 'MLPRegressor']
}
model_types = {
    'Random Forest': ['RandomForestClassifier', 'RandomForestRegressor'],
    'MLP': ['MLPClassifier', 'MLPRegressor']
}

def get_estimator_type(estimator_name):
    for estimator_type, estimator_names in estimator_types.items():
        if estimator_name in estimator_names:
            return estimator_type
    return None

def get_model_type(estimator_name):
    for model_type, estimator_names in model_types.items():
        if estimator_name in estimator_names:
            return model_type
    return None

def get_cv_results(grid_search_results):
    """
    Converts a dictionary of grid search results into a single dataframe.
    Includes additional information about the estimator and model type.
    """
    cv_results = pd.DataFrame()
    for dataset_type, grid_search in grid_search_results.items():
        # Add additional information to the dataframe
        current_cv_results = pd.DataFrame(grid_search.cv_results_).copy()
        current_cv_results['dataset_type'] = dataset_type
        current_cv_results['estimator_name'] = current_cv_results['param_estimator'].apply(lambda x: x.__class__.__name__)
        current_cv_results['estimator_type'] = current_cv_results['estimator_name'].apply(get_estimator_type)
        current_cv_results['model_type'] = current_cv_results['estimator_name'].apply(get_model_type)
        cv_results = pd.concat([cv_results, current_cv_results])
    # Reset the index
    cv_results.reset_index(inplace=True, drop=True)
    return cv_results
```

```python
# Get the results of the grid search and sort by mean test accuracy (descending)
cv_results = get_cv_results(grid_search_results)
processed_cv_results = cv_results[[
    'estimator_name',
    'estimator_type',
    'model_type',
    'dataset_type',
    'param_estimator',
    'params',
    'mean_train_accuracy',
    'std_train_accuracy',
    'mean_test_accuracy',
    'std_test_accuracy',
    'rank_test_accuracy' # For each type
]]
```

## Visualización con diagramas de violín
- Los diagramas de violín permiten visualizar la distribución de los datos. Estos diagramas muestran la distribución, es decir, que tan común es un dato (Yi, 2021).
- En el presente caso, esta distribución indica que tantos modelos están cercanos a un valor _X_ de la métrica de exactitud de validación, haciéndolos útiles para conocer donde se concentran la mayoría de modelos respecto a un algoritmo, tipo de modelo o dataset.
- Usando gráficos de violín, se puede visualizar con más exactitud donde están los valores de la distribución crean un gráfico de caja dentro de este, un gráfico de dispersión (scatter plot) o líneas. Para la visualización de los datos de este caso se usan líneas que indican donde están los puntos de datos, en este caso las exactitudes logradas por los modelos.
- Los gráficos de violín también permiten graficar dos distribuciones en el mismo violín, siendo ideal para comparaciones de grupos de datos, en este caso de modelos, algoritmos y datasets.
